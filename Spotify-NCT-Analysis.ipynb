```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df1 = pd.read_csv("Data/playlist_track_127.csv")
df2 = pd.read_csv("Data/playlist_track_dream.csv")
df3 = pd.read_csv("Data/playlist_track_wayv.csv")
df4 = pd.read_csv("Data/playlist_track_Uwish2dream.csv")
```

```python
df = pd.concat([df1, df2, df3, df4])
```

```python
df
```

```python
df.info()
```

```python
import pandas as pd

# Assuming 'df' is your DataFrame
df['duration_min'] = df['duration_ms'] / 60000
df['minutes'] = df['duration_min'].astype(int)
df['seconds'] = (df['duration_min'] - df['minutes']) * 60

# Convert seconds to string with leading zero padding
df['seconds_str'] = df['seconds'].astype(int).astype(str).str.zfill(2)

# Combine minutes and seconds with a colon (:)
df['duration_formatted'] = df['minutes'].astype(str) + ':' + df['seconds_str']
```

```python
# Optional: Drop the intermediate columns
df.drop(['duration_min', 'minutes', 'seconds', 'seconds_str'], axis=1, inplace=True)
```

```python
df
```

```python
df['year_of_release'] = pd.to_datetime(df['year_of_release'])
df['year'] = df['year_of_release'].dt.year
```

```python
df.drop(['year_of_release'], axis=1)
```

```python
df.to_csv("Data/playtrack.csv")
```

```python
def key_count():
    key_name = ["A", "A#", "B", "C", "C#", "D", "D#", "E",  "F", "F#", "G", "G#"]
    key_count = [len(df[df['key'] == i]) for i in range(12)]
    bars = plt.bar(key_name, key_count, color = "blue", width = 0.45)
    plt.xlabel('Key Signature')
    plt.ylabel('Count')
    plt.title('Key signature distribution')
    addvalue(bars = bars, is_time = False, fixed_precision = 0)
    plt.show()
```

```python
def addvalue(bars, is_time = False, fixed_precision = 3):
    for bar in bars:
        plt.text(bar.get_x() + bar.get_width() / 2, bar.get_height(), ("%d:%02d"%(bar.get_height() // 60, bar.get_height() % 60) if bar.get_height() >= 60 else ":%02d"%bar.get_height())if is_time else f"%.{fixed_precision}f"%bar.get_height(), ha = 'center', va = 'bottom', fontdict = {"size": 9})
```

```python
plt.figure(figsize=(12, 6))
sns.histplot(df['danceability'], bins=30, kde=True)
plt.title('Distribution of Danceability')
plt.xlabel('Danceability')
plt.ylabel('Frequency')
plt.show()
```

```python
# Mengurutkan DataFrame berdasarkan tempo secara descending dan mengambil 5 baris teratas
top_5_fastest = df[['Song', 'Artist', 'danceability']].sort_values(by='danceability', ascending=False).head(5)

# Membuat bar chart
plt.figure(figsize=(10, 6))
plt.barh(top_5_fastest['Song'] + ' - ' + top_5_fastest['Artist'], top_5_fastest['danceability'])
plt.xlabel('Danceability')
plt.ylabel('Lagu dan Artis')
plt.title('5 Lagu dengan Danceability Tercepat')
plt.gca().invert_yaxis()  # Membalik sumbu y agar lagu dengan tempo tercepat di atas
plt.show()
```
